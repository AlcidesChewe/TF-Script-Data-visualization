```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo para las visualizaciones
sns.set(style="whitegrid")
```

```python
# Cargar el dataset
df = pd.read_csv('amazon_review.csv')
```

```python
# Ver las primeras filas del dataset
print(df.head())

# Información general del dataset
print(df.info())

# Verificar valores nulos
print(df.isnull().sum())
```

```python
# Scatter plot para ver la relación entre los votos útiles y la calificación general
plt.figure(figsize=(10, 6))
sns.scatterplot(x='helpful_yes', y='overall', data=df)
plt.title('Relación entre Votos Útiles y Calificación General')
plt.xlabel('Votos Útiles')
plt.ylabel('Calificación General')
plt.savefig('votos_utiles_vs_calificacion.png')
plt.show()
```

```python
# Scatter plot para ver la relación entre la antigüedad de la reseña y el número de votos totales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='day_diff', y='total_vote', data=df)
plt.title('Relación entre Antigüedad de la Reseña y Número de Votos Totales')
plt.xlabel('Días desde la Reseña')
plt.ylabel('Número de Votos Totales')
plt.savefig('antiguedad_vs_votos_totales.png')
plt.show()
```

```python
# Scatter plot para ver la relación entre el número de votos útiles y la calificación general
plt.figure(figsize=(10, 6))
sns.scatterplot(x='overall', y='total_vote', data=df)
plt.title('Relación entre Calificación General y Número de Votos Totales')
plt.xlabel('Calificación General')
plt.ylabel('Número de Votos Totales')
plt.savefig('calificacion_vs_votos_totales.png')
plt.show()
```

```python
# Exportar datos limpios
df.to_csv('amazon_books_clean.csv', index=False)
```
